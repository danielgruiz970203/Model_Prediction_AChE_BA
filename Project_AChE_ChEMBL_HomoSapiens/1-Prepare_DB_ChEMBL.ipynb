# CREATE ".SDF" FILE WITH 3D COORDENATES AND SAVE

```python
Lista_3D = []
Lista_Smiles = []
for idx, mol in enumerate(suppl):  
    if mol is None: continue

    m2 = mol
    m3 = Chem.AddHs(m2)
    AllChem.EmbedMolecule(m3,randomSeed=0xf00d)
    
    Lista_Smiles.append(Chem.MolToSmiles(mol))
    
    Lista_3D.append('Molecule_' + str(idx))
    Lista_3D.append(Chem.MolToMolBlock(m3))
    Lista_3D.append("$$$$\n")
    
    
with open ('./inputCHEMBL_docking.sdf', 'w', encoding = 'UTF-8') as f:
    for x in Lista_3D:
        f.writelines(x)
        
Lista_3D    
```
