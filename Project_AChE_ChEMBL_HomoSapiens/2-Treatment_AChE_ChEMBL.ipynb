```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Descriptors, Lipinski
```

# RDKit treatment TO CALCULATE MOLECULAR DESCRIPTORS OF DOCKING FILE

```python
suppl_dock = Chem.SDMolSupplier('./output_docking_melhores_poses.sdf')
count = 0
suppl_dock_3D = [mol for mol in suppl_dock if mol is not None]

MolNames = [mol.GetProp("_Name") for mol in suppl_dock_3D]
print(len(MolNames))
suppl_dock_3D[0]
```

```python
suppl_dock_2D = [Chem.MolFromSmiles(Chem.MolToSmiles(mol)) for mol in suppl_dock_3D]
suppl_dock_2D[0]
```

```python
suppl_dock_smiles = [Chem.MolToSmiles(mol) for mol in suppl_dock if mol is not None]

print(len(suppl_dock_smiles))
suppl_dock_smiles[0]

```

```python
from mordred import Calculator, descriptors 
```

```python
calc = Calculator(descriptors, ignore_3D=False)
len(calc)
```

```python
df_2D_3D_des = calc.pandas(suppl_dock_3D)
df_2D_3D_des
```

```python
df_Molecules = pd.DataFrame()
df_Molecules["Mol_Name"] = MolNames
df_Molecules["Mol_#"] = [int(i.split("_")[-1].split(" ")[0]) for i in MolNames]
df_Molecules["Smiles"] = suppl_dock_smiles
df_Molecules['IC50_(nM)'] = range(len(suppl_dock_smiles))


df_Molecules
```

```python
IC50 = np.array(pd.read_csv("./IC50_file.tsv", sep = "\t") )
IC50
```

```python

count = 0
countint = 0
for tupl in IC50:
    count += 1
    mol = int(tupl[0].split("_")[-1].split(" ")[0])
    
    
    if mol in list(df_Molecules.iloc[:,1]):
        countint += 1
        index = list(df_Molecules.iloc[:,1]).index(mol)
        df_Molecules.iloc[index,3] = tupl[1]
        
df_Molecules['pIC50'] = -np.log10((1*10**(-9))*(df_Molecules.iloc[:,3]))        
print(count)
print(countint)
print(type(df_Molecules.iloc[0,3]))
df_Molecules
```

```python
df_Mol_des = pd.merge(df_Molecules, df_2D_3D_des, left_index=True, right_index=True)
df_Mol_des
```

# NOMALIZATION pIC50 and CLASSIFICATION OF SPECIES

```python
# normalização dos valores de pIC50 em uma escala de base 1
def norm_pIC50_col(data):
    norm=[]
    for elem in data["pIC50"]:
        if elem < 1:
            elem = 1
        if elem > 14:
            elem = 14
        norm.append(elem)
    data["pIC50_norm"] = norm
    x = data.drop("pIC50",1)
    return x

df_m_pIC50_norm = norm_pIC50_col(df_Mol_des)

##########################################################
#Descriçaõ da coluna nova de pIC50
print(df_m_pIC50_norm["pIC50_norm"].describe())

```

```python
#Classificação da bioatividade
Bioactivity_class = []
ativas = 0
inativas = 0
intermediate = 0
for i in df_m_pIC50_norm.loc[:,"pIC50_norm"]:
    if i >= df_m_pIC50_norm["pIC50_norm"].mean() + 0.75:
        Bioactivity_class.append("Active")
        ativas += 1
    elif  i <= df_m_pIC50_norm["pIC50_norm"].mean() - 0.75:
        Bioactivity_class.append("Inactive")
        inativas += 1
    else:
        Bioactivity_class.append("Intermediary")
        intermediate += 1
#Adição da coluna ao DF principal        
df_m_pIC50_norm["Bioactivity_class"] = Bioactivity_class
#Salvado do arquivo xlsx
df_m_pIC50_norm.to_csv("df_m_pIC50_norm.tsv", sep = "\t", encoding = "utf-8")
print("O numero de moleculas ativas é:  {}".format(ativas))
print("O numero de moleculas intermedias é:  {}".format(intermediate))
print("O numero de moleculas inativas é:  {}".format(inativas))
```

```python

#Geraçaõ da faixa de valores entre especies ativas e inativas para o nosso modelo 
df_m_pIC50_norm.to_csv('df_ML_complete.tsv', sep = '\t', index = False)
df_ML_2class = df_m_pIC50_norm[df_m_pIC50_norm.Bioactivity_class != "Intermediary"]
df_ML_2class.to_csv('df_ML_2class_complete.tsv', sep = '\t', index = False )
df_ML_2class
```
