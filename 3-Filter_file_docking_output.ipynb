```python
import pandas as pd
```

```python
df = pd.read_csv("../../output_docking_AChE/DockingResults-2.mvdresults", sep = "\t")
df.columns

No_poses = 3
```

```python
df["PoseEnergy"]
```

```python
df_ = pd.DataFrame()
for i in range(0,len(list(df.loc[:,"PoseEnergy"])),No_poses):
    
    if list(df.loc[:,"PoseEnergy"])[i] <= list(df.loc[:,"PoseEnergy"])[i+1] and list(df.loc[:,"PoseEnergy"])[i] <= list(df.loc[:,"PoseEnergy"])[i+2]:
               
        df_ = df_.append(pd.DataFrame(df.iloc[i,:]).T)
        
    elif list(df.loc[:,"PoseEnergy"])[i+1] <= list(df.loc[:,"PoseEnergy"])[i+2]:
        
        df_ = df_.append(pd.DataFrame(df.iloc[i+1,:]).T)
        
    else:
        if i+2 == 6968:
         
            df_ = df_.append(pd.DataFrame(df.iloc[i+1,:]).T)
        
        else:
        
            df_ = df_.append(pd.DataFrame(df.iloc[i+2,:]).T)

df_.to_csv("./Filter_file_docking_output.tsv", sep = "\t")
```
